```python
import pandas as pd
import pymongo
```

# Extract - Web Scraping

```python
#States to scrape from website
states = {

    'Georgia': 'GA',
    'Illinois': 'IL',
    'New York': 'NY',
}
```

```python
state_list=[]
for key, value in states.items():
    state = value.lower()
    state_list.append(state)
```

```python
tables_df=pd.DataFrame()
```

```python

table=[]
table_list=[]
for state in state_list:
    table=pd.read_html(f'https://www.bls.gov/oes/current/oes_{state}.htm')

    table[0]['state']=state
    table_list.append(table)
    
```

```python
table_list[1]
```

```python
table_list[2]
```

# Transform

```python
table_df=pd.DataFrame()
for table in table_list:
    tables_df=pd.concat([tables_df,table[0]])
```

```python
tables_df[tables_df['state']=='il']
```

```python
tables_df=tables_df.dropna().reset_index(drop=True)
tables_df=tables_df[tables_df['Level']=='major']
tables_df.head()
```

```python
tables_df['Occupation code']=tables_df['Occupation code'].apply(lambda x: x.replace('-','')).astype(float)
tables_df.head()
```

```python
cols=['Annual mean wage','Employment RSE','Median hourly wage','Mean hourly wage','Mean wage RSE','Employment','Employment per 1,000 jobs','Location quotient']
for col in cols:
    tables_df[col]=tables_df[col].apply(lambda x: x.replace('$','').replace('(','').replace('%','').replace(')','').replace(',','')).astype(float)
tables_df
```

```python
tables_states=tables_df.reset_index(drop=True)
tables_states.head()
```

```python
set(tables_states.state)
```

```python
states_df=tables_states.rename(columns={'Occupation title (click on the occupation title to view its profile)':'Occupation title'})
```

```python
tables_states.to_csv("ga_il_ny.csv", index=False, header=True)
```

```python
state_dict=states_df.to_dict(orient='records')
```
